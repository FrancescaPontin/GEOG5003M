## Option 2: write code to unzip file and save contents in a new folder

- Then read in the shp file from that folder

```python
# OPTION 2: Unzip the file, save and then read in the shp file

# Path to the uploaded zip file
zip_file_path = '/content/test_folder.zip'

# Directory to extract the contents to
extract_dir = '/content/test_folder_unzipped'

# Create the 'extract' directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# List the contents of the extracted directory
extracted_files = os.listdir(extract_dir)
print('Extracted files:', extracted_files)

```

```python
# read in the data, located within the extracted directory in the extracted 'test_folder'
thailand_2 =gpd.read_file(f'{extract_dir}/{extracted_files[0]}/tha_admbnda_adm0_rtsd_20220121.shp')
thailand_2.plot()
```
